```python
# importing nessasary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# getting the dataset into the pandas dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
# printing the data
print(raw_mail_data.loc[0])
```

```python
# knowing the shape of the dataset
raw_mail_data.shape
```

```python
# seeing if there are any null values
raw_mail_data.isnull().sum()
```

```python
# 
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# printing head of the dataset
mail_data.head(1)
```

```python
# substituting the ham and spam mail by 0 and 1
# 1 --> ham
# 0 --> spam
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
```

```python
# separating the label from message
X = mail_data['Message']
Y = mail_data['Category']
```

```python
# printing the message and labels
print(X)
print(Y)
```

```python
# dividing the data into training and testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,stratify = Y, random_state=2)
```

```python
# getting the shape of training, testing and raw mail data
print(X.shape, X_train.shape, X_test.shape)
```

```python
# changing the data for the model to read correctly and easily 
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase='true')
```

```python
# fitting the data 
X_train_features = feature_extraction.fit_transform(X_train)
# we are not fitting the testing data as we dont want to show the model the testing data
X_test_features = feature_extraction.transform(X_test)

# converting the y dataset from object to integer datatype
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
# now printing the new data set
print(X_test_features)
```

```python
# getting the logistic regression model
model = LogisticRegression()
```

```python
# fitting the training dataset into the model
model.fit(X_train_features, Y_train)
```

```python
# getting the accuracy of the model by using training data 
X_train_prediction = model.predict(X_train_features)
X_train_prediction_accuracy = accuracy_score(X_train_prediction, Y_train)
print('prediction of training data: ', X_train_prediction_accuracy)
```

```python
# getting the accuracy of the model by using testing data
X_test_prediction = model.predict(X_test_features)
X_test_prediction_accuracy = accuracy_score(X_test_prediction, Y_test)
print('prediction of testing data: ', X_test_prediction_accuracy)
```

```python
# getting the mail
input_data = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]
# 
input_data_transform = feature_extraction.transform(input_data)
# predicting the data using the model
prediction = model.predict(input_data_transform)
# printing the prediction
if(prediction == 0):
  print("it's a SPAM")
else:
  print("it's a HAM")
```
